```sql
elect*
from `workspace`.`default`.`bright_coffee_shop_case_study`;




SELECT min(transaction_time) AS min_time,
       max(transaction_time) AS max_date
from `workspace`.`default`.`bright_coffee_shop_case_study`;
     


---min date of data collection - 01/01/2023
SELECT min(transaction_date) AS min_date
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;


----max date of data collection - 30/06/2026 (period - 6 months)
SELECT max(transaction_date) AS max_date
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;


--- stores (Lower Manhattan, Hell's Kitchen, Astoria)
SELECT DISTINCT store_location 
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;

---products sold at the stores - 9 product categories
SELECT DISTINCT product_category
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;

----products details solf at the store 80 categories
SELECT DISTINCT product_detail
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;


SELECT DISTINCT product_type
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;


SELECT DISTINCT product_category AS category,
                product_detail AS product_name
FROM `workspace`.`default`.`bright_coffee_shop_case_study`;

---Missing values NO NULLS
SELECT*
FROM `workspace`.`default`.`bright_coffee_shop_case_study`
 WHERE unit_price IS NULL;

---lowest and highest unit price
SELECT min(unit_price) AS min_price,
       max(unit_price) AS max_price
from `workspace`.`default`.`bright_coffee_shop_case_study`;

 ---Extract day and month name
SELECT Dayname(transaction_date) AS Day_name,
        Monthname(transaction_date) AS Month_name,
        transaction_date
from `workspace`.`default`.`bright_coffee_shop_case_study`;

---Revenue calculation
SELECT unit_price,
        transaction_qty,
        unit_price* transaction_qty AS revenue
from `workspace`.`default`.`bright_coffee_shop_case_study`;
     
  ----Combining functions to get a clean enhanced data set
SELECT  
        transaction_id,
        transaction_date,
        transaction_time,
        transaction_qty,
        store_id,
        store_location,
        product_id,
        product_category,
        product_type,
---new columns to enhance data insight
        Dayname(transaction_date) AS Day_name,
        Monthname(transaction_date) AS Month_name,
        dayofmonth(transaction_date) AS Date_of_month,
---case statements day classification
      CASE
        WHEN Dayname(transaction_date) IN ('Sun', 'Sat') THEN 'weekend'
        ELSE 'weekday'
       END AS Day_classification,

---case statements time classification

       CASE
        WHEN date_format(transaction_time,'HH:MM:SS') BETWEEN '05:00:00' AND '08:59:59' THEN 'rush hour'
        WHEN date_format(transaction_time,'HH:MM:SS') BETWEEN '09:00:00' AND '11:59:59' THEN 'brunch'
        WHEN date_format(transaction_time,'HH:MM:SS') BETWEEN '12:00:00' AND '15:59:59' THEN 'afternoon'
        WHEN date_format(transaction_time,'HH:MM:SS') BETWEEN '16:00:00' AND '23:59:59' THEN 'evening'
        ELSE 'weekday'
       END AS Day_classification,
    ----spend buckets
     CASE
        WHEN (transaction_qty*unit_price) < 50  THEN '01.Low spender'
        WHEN (transaction_qty*unit_price) BETWEEN 51 AND 200  THEN '02. Medium spender'
        WHEN (transaction_qty*unit_price) BETWEEN 201 AND 300 THEN '03.High spender'
        ELSE '04.Blesser'
       END AS Spend_buckets,

       ---Add revenue
       transaction_qty*unit_price AS Revenue
         
FROM `workspace`.`default`.`bright_coffee_shop_case_study`; 
```
